```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../fairjob.csv')
```

```python
print(f"Total: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
df.sample(5)
```

```python
# we don't want id's of user , impression(session) and product(job) so we will drop those columns
df = df.drop(columns=['user_id','impression_id','product_id'])
```

```python
df
```

```python
# let's rename the feature 'protected_attribute' to 'gender'
df = df.rename(columns={'protected_attribute':'gender'})
```

```python
df.isnull().sum()
```

There are no missing values.

```python
sns.countplot(x='click', data=df)
plt.title('Click Distribution')
plt.show()
```

```python
print(f'Only {len(df[df['click']==1])/len(df)*100} % rows have click = 1')
```

Only 0.7 % datapoints are clicked rest are unclicked

```python
sns.countplot(x='gender', hue='click', data=df)
plt.title('Clicks by gender')
plt.show()
```

```python
print(f"Only {len(df[(df['click']==1) & (df['gender']==0)])/len(df[df['gender']==0])*100} % females have click = 1")
```

```python
print(f"Only {len(df[(df['click']==1) & (df['gender']==1)])/len(df[df['gender']==1])*100} % males have click = 1")
```

```python
sns.barplot(x='gender', y='click', data=df)
plt.title('Click Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
sns.barplot(x='senior', y='click', hue='gender', data=df)
plt.title('Click Rate by Senior')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop('click', axis=1).select_dtypes(include=[np.number])
y = df['click']
model = RandomForestClassifier(n_estimators=20, random_state=42)
model.fit(X, y)
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importances')
plt.show()
```

```python
sns.lineplot(x=df['rank'], y=df['click'])
plt.title('Rank distribution')
plt.show()
```
